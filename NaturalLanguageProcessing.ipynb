```python
txt = "The big cat ate the little mouse who was after the fresh cheese"
```

## Chunking

```python
txt_tokenized_postagged = nltk.pos_tag(word_tokenize(txt)) #Added POS Tags
```

```python
txt_tokenized_postagged
```

```python
grammar_np = r"NP:{<DT>?<JJ>*<NN>}"
```

```python
chunk_result = nltk.RegexpParser(grammar_np).parse(txt_tokenized_postagged)
```
